### Á∑¥Áøí 1.1: ÂâµÂª∫Á∞°ÂñÆÁ∑öÂúñ
ÂâµÂª∫‰∏ÄÂÄãÂåÖÂê´‰∏âÊ¢ùÁ∑öÁöÑÂúñË°®Ôºö
- sin(x)ÔºåÁî®Á¥ÖËâ≤ÂØ¶Á∑öË°®Á§∫
- cos(x)ÔºåÁî®ËóçËâ≤ËôõÁ∑öË°®Á§∫
- tan(x)ÔºåÁî®Á∂†Ëâ≤ÈªûÂäÉÁ∑öË°®Á§∫Ôºå‰ΩÜÈôêÂà∂yËª∏ÁØÑÂúçÂú®-2Âà∞2‰πãÈñì

Ë¶ÅÊ±ÇÊ∑ªÂä†Ê®ôÈ°å„ÄÅËª∏Ê®ôÁ±§„ÄÅÂúñ‰æãÂíåÁ∂≤Ê†º„ÄÇ

```python
# Ëß£Á≠î 1.1
x = np.linspace(0, 2*np.pi, 1000)  # ÂâµÂª∫Âæû0Âà∞2œÄÁöÑ1000ÂÄãÁ≠âÂàÜÈªû

plt.figure(figsize=(10, 6))  # Ë®≠ÁΩÆÂúñÂΩ¢Â§ßÂ∞è

# Áπ™Ë£Ω‰∏âÊ¢ù‰∏çÂêåÁöÑÁ∑ö
plt.plot(x, np.sin(x), 'r-', linewidth=2, label='sin(x)')       # Á¥ÖËâ≤ÂØ¶Á∑ö
plt.plot(x, np.cos(x), 'b--', linewidth=2, label='cos(x)')      # ËóçËâ≤ËôõÁ∑ö
plt.plot(x, np.tan(x), 'g-.', linewidth=2, label='tan(x)')      # Á∂†Ëâ≤ÈªûÂäÉÁ∑ö

# Ë®≠ÁΩÆyËª∏ÁØÑÂúçÈôêÂà∂
plt.ylim(-2, 2)

# Ê∑ªÂä†Ê®ôÈ°åÂíåËª∏Ê®ôÁ±§
plt.title('‰∏âËßíÂáΩÊï∏ÂúñË°®', fontsize=16)
plt.xlabel('x (ÂºßÂ∫¶)', fontsize=12)
plt.ylabel('y ÂÄº', fontsize=12)

# Ê∑ªÂä†Á∂≤Ê†ºÂíåÂúñ‰æã
plt.grid(True, alpha=0.3)
plt.legend(loc='upper right', fontsize=12)

plt.tight_layout()
plt.show()
```


### Á∑¥Áøí 1.2: ÂâµÂª∫Â≠êÂúñ
ÂâµÂª∫‰∏ÄÂÄã2Ë°å3ÂàóÁöÑÂ≠êÂúñÁ∂≤Ê†ºÔºåÂú®ÊØèÂÄãÂ≠êÂúñ‰∏≠Áπ™Ë£Ω‰∏çÂêåÈ°ûÂûãÁöÑÂúñË°®Ôºö
1. Ê≠£Âº¶Êõ≤Á∑ö
2. È§òÂº¶Êõ≤Á∑ö 
3. ÊåáÊï∏Êõ≤Á∑ö y = 2^x
4. Â∞çÊï∏Êõ≤Á∑ö y = log(x)
5. ÊããÁâ©Á∑ö y = x¬≤
6. Á´ãÊñπÊõ≤Á∑ö y = x¬≥

ÊØèÂÄãÂ≠êÂúñÈÉΩÊáâÊúâÈÅ©Áï∂ÁöÑÊ®ôÈ°å„ÄÇ

```python
# Ëß£Á≠î 1.2
# ÂâµÂª∫2Ë°å3ÂàóÁöÑÂ≠êÂúñÁ∂≤Ê†º
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Â∞áÂ≠êÂúñÈô£ÂàóÂ±ïÂπ≥Ôºå‰æøÊñºÂæ™Áí∞Ë®™Âïè
axes = axes.flatten()

# ÂâµÂª∫xÂÄºÈô£Âàó
x1 = np.linspace(0, 2*np.pi, 100)  # Áî®Êñº‰∏âËßíÂáΩÊï∏
x2 = np.linspace(0, 4, 100)         # Áî®ÊñºÊåáÊï∏ÂáΩÊï∏
x3 = np.linspace(0.1, 10, 100)      # Áî®ÊñºÂ∞çÊï∏ÂáΩÊï∏
x4 = np.linspace(-3, 3, 100)        # Áî®ÊñºÂÜ™ÂáΩÊï∏

# 1. Ê≠£Âº¶Êõ≤Á∑ö
axes[0].plot(x1, np.sin(x1), 'r-', linewidth=2)
axes[0].set_title('Ê≠£Âº¶ÂáΩÊï∏ (sin(x))', fontsize=14)
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].grid(True, alpha=0.3)

# 2. È§òÂº¶Êõ≤Á∑ö
axes[1].plot(x1, np.cos(x1), 'b-', linewidth=2)
axes[1].set_title('È§òÂº¶ÂáΩÊï∏ (cos(x))', fontsize=14)
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].grid(True, alpha=0.3)

# 3. ÊåáÊï∏Êõ≤Á∑ö y = 2^x
axes[2].plot(x2, 2**x2, 'g-', linewidth=2)
axes[2].set_title('ÊåáÊï∏ÂáΩÊï∏ (2^x)', fontsize=14)
axes[2].set_xlabel('x')
axes[2].set_ylabel('2^x')
axes[2].grid(True, alpha=0.3)

# 4. Â∞çÊï∏Êõ≤Á∑ö y = log(x)
axes[3].plot(x3, np.log(x3), 'm-', linewidth=2)
axes[3].set_title('Â∞çÊï∏ÂáΩÊï∏ (log(x))', fontsize=14)
axes[3].set_xlabel('x')
axes[3].set_ylabel('log(x)')
axes[3].grid(True, alpha=0.3)

# 5. ÊããÁâ©Á∑ö y = x¬≤
axes[4].plot(x4, x4**2, 'c-', linewidth=2)
axes[4].set_title('ÊããÁâ©Á∑ö (x¬≤)', fontsize=14)
axes[4].set_xlabel('x')
axes[4].set_ylabel('x¬≤')
axes[4].grid(True, alpha=0.3)

# 6. Á´ãÊñπÊõ≤Á∑ö y = x¬≥
axes[5].plot(x4, x4**3, 'y-', linewidth=2)
axes[5].set_title('Á´ãÊñπÊõ≤Á∑ö (x¬≥)', fontsize=14)
axes[5].set_xlabel('x')
axes[5].set_ylabel('x¬≥')
axes[5].grid(True, alpha=0.3)

# Ë™øÊï¥Â≠êÂúñÈñìË∑ù
plt.tight_layout()

# Ê∑ªÂä†Á∏ΩÊ®ôÈ°å
fig.suptitle('Êï∏Â≠∏ÂáΩÊï∏ÂúñË°®ÈõÜ', fontsize=20, y=1.02)
plt.show()
```


### Á∑¥Áøí 1.3: Èù¢ÂêëÂ∞çË±°È¢®Ê†ºÁπ™Âúñ

- ÂâµÂª∫50ÂÄãÈö®Ê©üÈªû
- ÈªûÁöÑÂ§ßÂ∞èÊáâË©≤Èö®Ê©üËÆäÂåñ

```python
# Ëß£Á≠î 1.3
# ÂâµÂª∫Êï∏Êìö
n_points = 50
x = np.random.rand(n_points) * 10  # ÁØÑÂúç 0-10 ÁöÑÈö®Ê©ü x ÂÄº
y = np.random.rand(n_points) * 10  # ÁØÑÂúç 0-10 ÁöÑÈö®Ê©ü y ÂÄº
sizes = np.random.rand(n_points) * 300 + 50  # ÁØÑÂúç 50-350 ÁöÑÈö®Ê©üÈªûÂ§ßÂ∞è
colors = np.random.rand(n_points)  # ÁØÑÂúç 0-1 ÁöÑÈö®Ê©üÂÄºÁî®ÊñºÈ°èËâ≤Êò†Â∞Ñ

# ÂâµÂª∫ÂúñÂΩ¢ÂíåËª∏Â∞çË±°
fig, ax = plt.subplots(figsize=(10, 8))

# ‰ΩøÁî®Êï£ÈªûÂúñÂáΩÊï∏
scatter = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', 
                    alpha=0.7, edgecolors='black', linewidths=0.5)

# Ê∑ªÂä†Ê®ôÈ°åÂíåËª∏Ê®ôÁ±§
ax.set_title('Èö®Ê©üÈªûÊï£ÈªûÂúñ', fontsize=16)
ax.set_xlabel('X Ëª∏', fontsize=14)
ax.set_ylabel('Y Ëª∏', fontsize=14)

# Ê∑ªÂä†Á∂≤Ê†º
ax.grid(True, linestyle='--', alpha=0.6)

# Ë®≠ÁΩÆËª∏ÁØÑÂúç
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Ê∑ªÂä†È°èËâ≤Ê¢ù
cbar = fig.colorbar(scatter, ax=ax, label='È°èËâ≤ÂÄº')
cbar.set_label('È°èËâ≤ÂÄº', fontsize=12)

# Ê∑ªÂä†È°çÂ§ñÁöÑÁæéÂåñÂÖÉÁ¥†
ax.set_facecolor('#f8f9fa')  # Ë®≠ÁΩÆËÉåÊôØÈ°èËâ≤
for spine in ax.spines.values():
    spine.set_linewidth(1.5)  # Âä†Á≤óÈÇäÊ°ÜÁ∑ö

plt.tight_layout()
plt.show()
```

### Á∑¥Áøí 2.1: ÂâµÂª∫‰∏¶ÂàÜÊûêÊï∏ÊìöÈõÜ
ÂâµÂª∫‰∏ÄÂÄãÊ®°Êì¨ÁöÑÂ≠∏ÁîüÊàêÁ∏æÊï∏ÊìöÈõÜÔºåÂåÖÂê´‰ª•‰∏ãÂàóÔºö
- ÂßìÂêç (Èö®Ê©üÁîüÊàêÔºåÂ¶Ç 'Student_1', 'Student_2'...)
- Áè≠Á¥ö (Èö®Ê©üÂàÜÈÖçÔºåÂ¶Ç 'Class A', 'Class B', 'Class C')
- Êï∏Â≠∏ÂàÜÊï∏ (ÁîüÊàê0-100‰πãÈñìÁöÑÈö®Ê©üÂàÜÊï∏)
- ÁßëÂ≠∏ÂàÜÊï∏ (ÁîüÊàê0-100‰πãÈñìÁöÑÈö®Ê©üÂàÜÊï∏)
- Ë™ûÊñáÂàÜÊï∏ (ÁîüÊàê0-100‰πãÈñìÁöÑÈö®Ê©üÂàÜÊï∏)
- ÈÄöÈÅéÊÉÖÊ≥Å (Ëã•‰∏âÁßëÂπ≥ÂùáÂàÜ‚â•60ÔºåÂâáÁÇ∫'Pass'ÔºåÂê¶ÂâáÁÇ∫'Fail')

ÂâµÂª∫‰∏ÄÂÄãÂåÖÂê´30ÂêçÂ≠∏ÁîüÁöÑÊï∏ÊìöÈõÜ„ÄÇ

```python
# Ëß£Á≠î 2.1
# Ë®≠ÁΩÆÂ≠∏ÁîüÊï∏Èáè
n_students = 30

# ÂâµÂª∫ÂêÑÂàóÊï∏Êìö
student_names = [f'Student_{i+1}' for i in range(n_students)]
classes = np.random.choice(['Class A', 'Class B', 'Class C'], size=n_students)
math_scores = np.random.randint(0, 101, size=n_students)  # 0-100‰πãÈñìÁöÑÈö®Ê©üÂàÜÊï∏
science_scores = np.random.randint(0, 101, size=n_students)
language_scores = np.random.randint(0, 101, size=n_students)

# Ë®àÁÆóÂπ≥ÂùáÂàÜ
average_scores = (math_scores + science_scores + language_scores) / 3

# Âà§Êñ∑ÈÄöÈÅéÊÉÖÊ≥Å
pass_status = ['Pass' if avg >= 60 else 'Fail' for avg in average_scores]

# ÂâµÂª∫DataFrame
student_data = pd.DataFrame({
    'ÂßìÂêç': student_names,
    'Áè≠Á¥ö': classes,
    'Êï∏Â≠∏': math_scores,
    'ÁßëÂ≠∏': science_scores,
    'Ë™ûÊñá': language_scores,
    'Âπ≥ÂùáÂàÜ': average_scores,
    'ÈÄöÈÅéÊÉÖÊ≥Å': pass_status
})

# È°ØÁ§∫Êï∏Êìö
print("Â≠∏ÁîüÊàêÁ∏æÊï∏ÊìöÈõÜÈ†êË¶Ω:")
print(student_data.head(10))

# È°ØÁ§∫Âü∫Êú¨Áµ±Ë®àË≥áË®ä
print("\nÂü∫Êú¨Áµ±Ë®àË≥áË®ä:")
print(student_data[['Êï∏Â≠∏', 'ÁßëÂ≠∏', 'Ë™ûÊñá', 'Âπ≥ÂùáÂàÜ']].describe())

# È°ØÁ§∫ÈÄöÈÅéÁéá
pass_rate = student_data['ÈÄöÈÅéÊÉÖÊ≥Å'].value_counts(normalize=True)
print("\nÈÄöÈÅéÁéá:")
print(pass_rate)

# È°ØÁ§∫ÂêÑÁè≠Á¥ö‰∫∫Êï∏
class_counts = student_data['Áè≠Á¥ö'].value_counts()
print("\nÂêÑÁè≠Á¥ö‰∫∫Êï∏:")
print(class_counts)
```


### Á∑¥Áøí 2.2: Áπ™Ë£ΩÈï∑Ê¢ùÂúñÊØîËºÉ‰∏çÂêåÁè≠Á¥öÁöÑÂπ≥ÂùáÂàÜÊï∏
Ë®àÁÆóÊØèÂÄãÁè≠Á¥öÂú®‰∏âÂÄãÁßëÁõÆ‰∏äÁöÑÂπ≥ÂùáÂàÜÊï∏Ôºå‰∏¶ÂâµÂª∫‰∏ÄÂÄãÂàÜÁµÑÈï∑Ê¢ùÂúñÈÄ≤Ë°åÊØîËºÉ„ÄÇ
Ë¶ÅÊ±ÇÊ∑ªÂä†ÈÅ©Áï∂ÁöÑÊ®ôÈ°å„ÄÅËª∏Ê®ôÁ±§ÂíåÂúñ‰æã„ÄÇ

```python
# Ëß£Á≠î 2.2
# Ë®àÁÆóÊØèÂÄãÁè≠Á¥öÂú®ÂêÑÁßëÁõÆÁöÑÂπ≥ÂùáÂàÜÊï∏
class_averages = student_data.groupby('Áè≠Á¥ö')[['Êï∏Â≠∏', 'ÁßëÂ≠∏', 'Ë™ûÊñá']].mean().reset_index()

# ÈáçÂ°ëÊï∏Êìö‰ª•‰æøÊñºÁπ™Âúñ
class_averages_melted = pd.melt(class_averages, 
                               id_vars='Áè≠Á¥ö',
                               value_vars=['Êï∏Â≠∏', 'ÁßëÂ≠∏', 'Ë™ûÊñá'],
                               var_name='ÁßëÁõÆ',
                               value_name='Âπ≥ÂùáÂàÜÊï∏')

print("ÈáçÂ°ëÂæåÁöÑÊï∏Êìö:")
print(class_averages_melted)

# Ë®≠ÁΩÆÈ°èËâ≤
colors = ['#3498db', '#2ecc71', '#e74c3c']

# ÂâµÂª∫ÂàÜÁµÑÈï∑Ê¢ùÂúñ
plt.figure(figsize=(12, 7))

# Ë®àÁÆóÈï∑Ê¢ùÁöÑ‰ΩçÁΩÆ
classes = class_averages['Áè≠Á¥ö'].unique()
subjects = ['Êï∏Â≠∏', 'ÁßëÂ≠∏', 'Ë™ûÊñá']
n_classes = len(classes)
n_subjects = len(subjects)
width = 0.25  # Èï∑Ê¢ùÂØ¨Â∫¶
indices = np.arange(n_classes)  # Áè≠Á¥ö‰ΩçÁΩÆ

# Áπ™Ë£ΩÈï∑Ê¢ùÂúñ
for i, subject in enumerate(subjects):
    # Áç≤ÂèñÁâπÂÆöÁßëÁõÆÊØèÂÄãÁè≠Á¥öÁöÑÂπ≥ÂùáÂàÜ
    scores = class_averages[subject].values
    
    # Áπ™Ë£ΩÈï∑Ê¢ù
    bars = plt.bar(indices + (i - 1) * width, scores, width, 
                  label=subject, color=colors[i], edgecolor='black', linewidth=1)
    
    # Ê∑ªÂä†Êï∏ÂÄºÊ®ôÁ±§
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 1,
                f'{height:.1f}', ha='center', va='bottom', fontsize=10)

# Ê∑ªÂä†Ê®ôÈ°åÂíåËª∏Ê®ôÁ±§
plt.title('ÂêÑÁè≠Á¥öÂú®‰∏çÂêåÁßëÁõÆÁöÑÂπ≥ÂùáÂàÜÊï∏ÊØîËºÉ', fontsize=16, pad=15)
plt.xlabel('Áè≠Á¥ö', fontsize=14, labelpad=10)
plt.ylabel('Âπ≥ÂùáÂàÜÊï∏', fontsize=14, labelpad=10)

# Ë®≠ÁΩÆxËª∏ÂàªÂ∫¶Ê®ôÁ±§
plt.xticks(indices, classes, fontsize=12)

# Ê∑ªÂä†Á∂≤Ê†ºÁ∑ö
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ê∑ªÂä†Âúñ‰æã
plt.legend(fontsize=12, title='ÁßëÁõÆ', title_fontsize=13)

# Ë™øÊï¥‰ΩàÂ±Ä
plt.tight_layout()

# È°ØÁ§∫ÂúñÂΩ¢
plt.show()
```


### Á∑¥Áøí 2.4: ÂâµÂª∫Êï£ÈªûÁü©Èô£
‰ΩøÁî®pandasÁöÑscatter_matrixÂáΩÊï∏ÔºåÂâµÂª∫‰∏ÄÂÄã‰∏âÁßëÊàêÁ∏æÁöÑÊï£ÈªûÁü©Èô£ÂúñÔºå
‰ΩøÁî®"ÈÄöÈÅéÊÉÖÊ≥Å"‰ΩúÁÇ∫ÈªûÁöÑÈ°èËâ≤ÂçÄÂàÜ„ÄÇ

```python
# Ëß£Á≠î 2.4
# ÂæûpandasÂ∞éÂÖ•scatter_matrix (Â¶ÇÊûúÈúÄË¶Å)
from pandas.plotting import scatter_matrix

# ÂâµÂª∫Êï£ÈªûÁü©Èô£
# Âè™ÈÅ∏ÊìáÈúÄË¶ÅÁöÑÂàóÔºöÊï∏Â≠∏„ÄÅÁßëÂ≠∏„ÄÅË™ûÊñá
columns_to_plot = ['Êï∏Â≠∏', 'ÁßëÂ≠∏', 'Ë™ûÊñá']

# Âü∫Êñº"ÈÄöÈÅéÊÉÖÊ≥Å"ÂâµÂª∫È°èËâ≤Êò†Â∞Ñ
colors = {'Pass': 'green', 'Fail': 'red'}
color_values = [colors[status] for status in student_data['ÈÄöÈÅéÊÉÖÊ≥Å']]

# Áπ™Ë£ΩÊï£ÈªûÁü©Èô£
fig, axes = plt.subplots(figsize=(12, 10))
scatter_matrix = pd.plotting.scatter_matrix(
    student_data[columns_to_plot], 
    figsize=(12, 10),
    diagonal='kde',  # Â∞çËßíÁ∑ö‰∏äÈ°ØÁ§∫Ê†∏ÂØÜÂ∫¶‰º∞Ë®à
    marker='o',
    s=80,  # ÈªûÁöÑÂ§ßÂ∞è
    alpha=0.8,  # ÈÄèÊòéÂ∫¶
    c=color_values,  # Âü∫ÊñºÈÄöÈÅéÁãÄÊÖãË®≠ÁΩÆÈ°èËâ≤
    hist_kwds={'bins': 15},  # Áõ¥ÊñπÂúñÁöÑbinsÊï∏
    range_padding=0.1  # ÂùêÊ®ôËª∏ÁØÑÂúçÁöÑÂ°´ÂÖÖ
)

# ÁæéÂåñÊï£ÈªûÁü©Èô£
for ax in scatter_matrix.flatten():
    ax.xaxis.label.set_rotation(45)
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha('right')
    
    # Ë®≠ÁΩÆÁ∂≤Ê†º
    ax.grid(True, linestyle='--', alpha=0.3)
    
    # ÁæéÂåñÂàªÂ∫¶Ê®ôÁ±§
    ax.tick_params(axis='both', labelsize=9)

# Ê∑ªÂä†Âúñ‰æãÔºàÊâãÂãïÊ∑ªÂä†ÔºåÂõ†ÁÇ∫scatter_matrixÊ≤íÊúâÁõ¥Êé•ÊîØÊåÅÂúñ‰æãÔºâ
import matplotlib.patches as mpatches
pass_patch = mpatches.Patch(color='green', label='ÈÄöÈÅé')
fail_patch = mpatches.Patch(color='red', label='‰∏çÈÄöÈÅé')
plt.legend(handles=[pass_patch, fail_patch], 
           loc='upper right', 
           bbox_to_anchor=(0.95, 0.95),
           fontsize=12)

# Ê∑ªÂä†Á∏ΩÊ®ôÈ°å
plt.suptitle('Â≠∏ÁîüÊàêÁ∏æÊï£ÈªûÁü©Èô£ (ÊåâÈÄöÈÅéÊÉÖÊ≥ÅÊ®ôËâ≤)', fontsize=16, y=0.98)

# Ë™øÊï¥‰ΩàÂ±Ä
plt.tight_layout()
plt.subplots_adjust(top=0.95)

# È°ØÁ§∫ÂúñÂΩ¢
plt.show()
```

‰ΩøÁî®TitanicÊï∏ÊìöÈõÜÔºåÂâµÂª∫‰∏ÄÂÄãÂ∞èÊèêÁê¥ÂúñÊØîËºÉÂ≠òÊ¥ª(survived)ËàáÈùûÂ≠òÊ¥ª‰πòÂÆ¢ÁöÑÂπ¥ÈΩ°ÂàÜ‰Ωà„ÄÇ

```python
# Ëß£Á≠î 4.3
# Â∞éÂÖ•GridSpec
from matplotlib.gridspec import GridSpec

# ÂâµÂª∫‰∏ÄÂÄãËºÉÂ§ßÁöÑÂúñÂΩ¢
plt.figure(figsize=(15, 12))

# ‰ΩøÁî®GridSpecÂâµÂª∫ÈùûÂùáÂãª‰ΩàÂ±Ä
gs = GridSpec(3, 3, width_ratios=[1.5, 1, 1], height_ratios=[1, 1.5, 1], 
             wspace=0.4, hspace=0.4)

# 1. ÊäòÁ∑öÂúñ (Â∑¶‰∏äÔºåË∑®Ë∂äÂÖ©Âàó)
ax1 = plt.subplot(gs[0, 0:2])

# ÂâµÂª∫‰∏¶Áπ™Ë£ΩË°∞Ê∏õÊ≠£Âº¶Êõ≤Á∑ö
x_line = np.linspace(0, 10, 200)
y_line = np.sin(x_line) * np.exp(-x_line/5)
ax1.plot(x_line, y_line, color='blue', linewidth=2)
ax1.set_title('Ë°∞Ê∏õÊ≠£Âº¶Ê≥¢', fontsize=14)
ax1.set_xlabel('x Ëª∏', fontsize=12)
ax1.set_ylabel('y ÂÄº', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)

# 2. Êï£ÈªûÂúñ (Âè≥‰∏ä)
ax2 = plt.subplot(gs[0, 2])

# ÂâµÂª∫Èö®Ê©üÊï£Èªû
n = 50
x_scatter = np.random.rand(n) * 10
y_scatter = np.random.rand(n) * 10
colors = np.random.rand(n)
sizes = np.random.rand(n) * 200 + 50

scatter = ax2.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.6, 
                     cmap='viridis', edgecolor='black')
ax2.set_title('Èö®Ê©üÊï£Èªû', fontsize=14)
ax2.set_xlabel('x Ëª∏', fontsize=12)
ax2.set_ylabel('y Ëª∏', fontsize=12)
ax2.grid(True, alpha=0.3)

# 3. ÁÜ±Âúñ (‰∏≠ÈñìÂ∑¶ÔºåËºÉÂ§ß)
ax3 = plt.subplot(gs[1, 0])

# ÂâµÂª∫‰∏ÄÂÄãÁõ∏ÈóúÁü©Èô£È¢®Ê†ºÁöÑÁÜ±Âúñ
data = np.corrcoef(np.random.rand(10, 10))
im = ax3.imshow(data, cmap='coolwarm', interpolation='nearest', vmin=-1, vmax=1)
ax3.set_title('Áõ∏ÈóúÁü©Èô£ÁÜ±Âúñ', fontsize=14)
plt.colorbar(im, ax=ax3, fraction=0.046, pad=0.04)
ax3.set_xticks(np.arange(data.shape[1]))
ax3.set_yticks(np.arange(data.shape[0]))
ax3.set_xticklabels([f'X{i+1}' for i in range(data.shape[1])], fontsize=8, rotation=45)
ax3.set_yticklabels([f'X{i+1}' for i in range(data.shape[0])], fontsize=8)

# 4. Áõ¥ÊñπÂúñ (‰∏≠Èñì)
ax4 = plt.subplot(gs[1, 1])

# ÂâµÂª∫Ê≠£ÊÖãÂàÜÂ∏ÉÁöÑÁõ¥ÊñπÂúñ
data_hist = np.random.normal(0, 1, 1000)
ax4.hist(data_hist, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
ax4.set_title('Ê≠£ÊÖãÂàÜÂ∏ÉÁõ¥ÊñπÂúñ', fontsize=14)
ax4.set_xlabel('ÂÄº', fontsize=12)
ax4.set_ylabel('È†ªÁéá', fontsize=12)
ax4.grid(True, axis='y', alpha=0.3)

# 5. È§ÖÂúñ (‰∏≠ÈñìÂè≥)
ax5 = plt.subplot(gs[1, 2])

# ÂâµÂª∫È§ÖÂúñ
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [25, 30, 15, 10, 20]
colors = plt.cm.Paired(np.arange(len(labels))/len(labels))
wedges, texts, autotexts = ax5.pie(sizes, labels=labels, autopct='%1.1f%%',
                                  startangle=90, colors=colors, 
                                  wedgeprops={'width': 0.5, 'edgecolor': 'w'},
                                  textprops={'fontsize': 10})
plt.setp(autotexts, size=10, weight="bold")
ax5.set_title('Áí∞ÂΩ¢Âúñ', fontsize=14)
ax5.axis('equal')  # Á¢∫‰øùÈ§ÖÂúñÊòØÂúìÁöÑ

# 6. Á≠âÈ´òÁ∑öÂúñ (Â∫ïÈÉ®ÔºåË∑®Ë∂ä‰∏âÂàó)
ax6 = plt.subplot(gs[2, :])

# ÂâµÂª∫Á≠âÈ´òÁ∑öÂúñ
x_contour = np.linspace(-3, 3, 100)
y_contour = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x_contour, y_contour)
Z = np.sin(X) * np.cos(Y)

contour = ax6.contourf(X, Y, Z, 20, cmap='viridis')
ax6.contour(X, Y, Z, 10, colors='white', alpha=0.5, linewidths=0.5)
ax6.set_title('Á≠âÈ´òÁ∑öÂúñ', fontsize=14)
ax6.set_xlabel('x Ëª∏', fontsize=12)
ax6.set_ylabel('y Ëª∏', fontsize=12)
plt.colorbar(contour, ax=ax6, shrink=0.6, aspect=20, pad=0.02)

# Ê∑ªÂä†Êï¥È´îÊ®ôÈ°å
plt.suptitle('Â§öÊ®£ÂåñÂúñË°®‰ΩàÂ±ÄÂ±ïÁ§∫', fontsize=20, y=0.98)

# Ë™øÊï¥‰ΩàÂ±Ä
plt.tight_layout()
plt.subplots_adjust(top=0.93)

# È°ØÁ§∫ÂúñÂΩ¢
plt.show()
```
