## Part A — 統計學是什麼？

### 統計學在 DA / DS / DE 的角色

| 角色 | 核心任務 | 統計學的用途 |
|:-----|:---------|:-------------|
| **資料分析師 (DA)** | 報表、儀表板、業務洞察 | 描述統計、假設檢定、A/B Testing |
| **資料科學家 (DS)** | 建模、預測、實驗設計 | 機率分佈、迴歸分析、貝氏推論 |
| **資料工程師 (DE)** | 管線建置、資料品質 | 異常偵測、資料驗證、抽樣策略 |

> 不論哪個角色，**統計學都是資料從業者的共同語言**。

### 描述統計 vs 推論統計

| 面向 | 描述統計 (Descriptive) | 推論統計 (Inferential) |
|:-----|:----------------------|:----------------------|
| **目的** | 整理、摘要已有資料 | 從樣本推論母體特性 |
| **工具** | 平均數、中位數、標準差、圖表 | 假設檢定、信賴區間、迴歸 |
| **不確定性** | 不涉及 | 量化不確定性（p-value、信賴水準） |
| **範例** | 「這 891 位乘客的平均年齡是 29.7 歲」 | 「全體乘客的平均年齡介於 28.5–30.9 歲（95% 信賴區間）」 |

本課程的 13 章將**從描述統計出發，逐步深入推論統計**。

### 資料驅動決策流程 (Data-Driven Decision Making)

```
商業問題 → 資料蒐集 → 資料清理 → 探索性分析 (EDA)
    → 統計建模 / 假設檢定 → 結論與決策 → 追蹤與迭代
```

統計學貫穿整個流程：
- **資料蒐集**階段：抽樣方法 (Sampling Methods) 決定樣本品質
- **EDA 階段**：描述統計量幫助理解資料全貌
- **建模階段**：推論統計提供決策的數學基礎
- **決策階段**：統計顯著性 (Statistical Significance) 判斷結果是否可信

### 💡 知識補給站 — 統計學的兩大陣營

| 學派 | 頻率學派 (Frequentist) | 貝氏學派 (Bayesian) |
|:-----|:----------------------|:-------------------|
| **核心觀點** | 機率 = 長期頻率 | 機率 = 信念程度 |
| **參數觀** | 參數是固定但未知的常數 | 參數是隨機變數，有先驗分佈 |
| **代表方法** | p-value、信賴區間 | 後驗分佈、貝氏因子 |
| **優勢** | 計算簡單、廣泛使用 | 可融入先驗知識、小樣本表現佳 |
| **本課程** | Ch02–Ch10 主要採用 | Ch11–Ch13 進階介紹 |

> 兩派並非對立，而是**看待不確定性的不同視角**。實務中常搭配使用。

## Part B — 母體與樣本

### 母體 (Population) vs 樣本 (Sample)

| 概念 | 母體 (Population) | 樣本 (Sample) |
|:-----|:------------------|:--------------|
| **定義** | 研究對象的全體集合 | 從母體中抽取的子集 |
| **大小** | 通常龐大或無限 | 有限且可操作 |
| **數值** | 參數 (Parameter)，如 μ、σ | 統計量 (Statistic)，如 x̄、s |
| **符號** | 希臘字母（μ, σ, π） | 拉丁字母（x̄, s, p̂） |
| **Titanic** | 所有 2,224 名乘客 | Kaggle 訓練集的 891 名乘客 |

### 為什麼要抽樣？

1. **成本考量**：普查 (Census) 耗時耗力，抽樣更經濟
2. **可行性**：有些母體根本無法全部觀測（如「所有未來的顧客」）
3. **破壞性測試**：如燈泡壽命測試，不可能測完所有產品
4. **時效性**：抽樣可以更快得到結果

> 關鍵不在樣本大小，而在**樣本是否具代表性 (Representativeness)**。

## Part C — 資料類型與測量尺度

### 四種測量尺度 (Measurement Scales)

| 尺度 | 英文 | 特性 | Titanic 範例 | 可做的運算 |
|:-----|:-----|:-----|:-------------|:-----------|
| **名目** | Nominal | 分類、無順序 | `Sex`（male/female）、`Embarked`（S/C/Q） | = ≠ |
| **順序** | Ordinal | 分類、有順序 | `Pclass`（1st > 2nd > 3rd） | = ≠ > < |
| **區間** | Interval | 等距、無真正零點 | 溫度（攝氏）、日期 | = ≠ > < + − |
| **比率** | Ratio | 等距、有真正零點 | `Age`、`Fare`、`SibSp` | = ≠ > < + − × ÷ |

> **記憶口訣：** 名順區比（NOIR — Nominal, Ordinal, Interval, Ratio）

### 質性 vs 量性、離散 vs 連續

```
資料類型
├── 質性 (Qualitative / Categorical)
│   ├── 名目 (Nominal)：Sex, Embarked
│   └── 順序 (Ordinal)：Pclass
└── 量性 (Quantitative / Numerical)
    ├── 離散 (Discrete)：SibSp, Parch, Survived
    └── 連續 (Continuous)：Age, Fare
```

| 類型 | 質性 (Categorical) | 量性 (Numerical) |
|:-----|:-------------------|:-----------------|
| **定義** | 描述類別或屬性 | 描述數量或程度 |
| **運算** | 計數、眾數 | 平均、標準差、中位數 |
| **視覺化** | 長條圖、圓餅圖 | 直方圖、箱型圖、散佈圖 |

### 💡 知識補給站 — 資料類型決定分析方法

| 自變數 (X) | 因變數 (Y) | 適用方法 |
|:-----------|:-----------|:---------|
| 類別 | 類別 | 卡方檢定 (Chi-squared Test) |
| 類別 | 連續 | t 檢定 / ANOVA |
| 連續 | 連續 | 相關分析 / 迴歸分析 |
| 連續 | 類別 | 邏輯迴歸 (Logistic Regression) |

> 在後續章節中，我們會依序學習這些方法。**正確辨識資料類型是選擇分析方法的第一步。**

## Part D — 資料品質初探

### 常見的資料品質問題

| 問題 | 說明 | Titanic 範例 |
|:-----|:-----|:-------------|
| **缺值 (Missing Values)** | 欄位沒有值 | `Age`（177 筆）、`Cabin`（687 筆） |
| **離群值 (Outliers)** | 極端偏離的數值 | `Fare` 中有 $512.33 的票價 |
| **型別錯誤 (Type Errors)** | 數值存成字串，或類別存成數字 | `Pclass` 看似數字，實為順序類別 |

> 資料品質是統計分析的基礎。垃圾進、垃圾出 (Garbage In, Garbage Out, GIGO)。

````python
df.isnull().sum()
````

````python
df.describe()
````

### 口訣

> **資料品質三步驟：看缺值、看分佈、看型別**
>
> - `isnull().sum()`：哪些欄位有缺值？嚴重嗎？
> - `describe()`：數值欄位的分佈合理嗎？有沒有離群值？
> - `dtypes` + `value_counts()`：型別正確嗎？類別欄位有沒有打字錯誤？
