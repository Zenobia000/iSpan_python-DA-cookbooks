## 1. 什麼是多型？

多型的意思是：**同樣的操作，不同的物件會有不同的行為。**

你已經在用多型了，只是可能沒注意到：

```python
# len() 對不同型別有不同行為 — 這就是多型
print(len("Hello"))    # 5（字串的長度）
print(len([1, 2, 3]))  # 3（串列的長度）
```

在 OOP 中，多型讓你可以用同一個介面操作不同的物件：

```python
class Dog:
    def speak(self):
        return "汪汪！"

class Cat:
    def speak(self):
        return "喵～"

class Duck:
    def speak(self):
        return "呱呱！"

# 同一個函式，處理不同的物件
def animal_sound(animal):
    print(animal.speak())

for animal in [Dog(), Cat(), Duck()]:
    animal_sound(animal)    # 各自呼叫自己的 speak()
```

## 2. 鴨子型別（Duck Typing）

> 「如果牠走路像鴨子、叫聲像鴨子，那牠就是鴨子。」

Python 不在乎物件的型別，只在乎它**有沒有你要用的方法**：

```python
class Robot:
    """Robot 不是動物，但也有 speak() 方法"""
    def speak(self):
        return "嗶嗶嗶..."

# 完全可以丟進 animal_sound，因為它有 speak()
animal_sound(Robot())  # 嗶嗶嗶...
```

鴨子型別的好處是**彈性高**，但缺點是如果忘了實作某個方法，要到**執行時才會發現錯誤**。

## 3. 抽象類別（ABC）— 強制規範介面

如果你想確保子類別一定要實作某些方法，可以使用抽象類別：

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    """抽象類別：定義所有形狀必須有的方法"""

    @abstractmethod
    def area(self):
        """計算面積（子類別必須實作）"""
        pass

    @abstractmethod
    def perimeter(self):
        """計算周長（子類別必須實作）"""
        pass

    def describe(self):
        """一般方法：子類別可以直接使用"""
        return f"面積={self.area():.2f}, 周長={self.perimeter():.2f}"
```

```python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14159 * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
```

```python
shapes = [Circle(5), Rectangle(3, 4)]

for s in shapes:
    print(s.describe())
# 面積=78.54, 周長=31.42
# 面積=12.00, 周長=14.00
```

### 抽象類別的保護作用

```python
# 無法直接實例化抽象類別
try:
    shape = Shape()  # TypeError: Can't instantiate abstract class
except TypeError as e:
    print(f"錯誤: {e}")

# 如果子類別忘了實作抽象方法，也會報錯
class Triangle(Shape):
    def area(self):
        return 0
    # 忘了實作 perimeter()...

try:
    t = Triangle()  # TypeError: Can't instantiate abstract class Triangle
                    # with abstract method perimeter
except TypeError as e:
    print(f"錯誤: {e}")
```

錯誤在**建立物件時就會發生**，不用等到呼叫方法時才炸掉。

## 4. ABC vs 鴨子型別：怎麼選？

| 考量 | 鴨子型別 | ABC 抽象類別 |
|---|---|---|
| 彈性 | 高，任何物件都能用 | 中，必須繼承 ABC |
| 安全性 | 低，執行時才發現缺方法 | 高，建立物件時就檢查 |
| 適合場景 | 小型專案、快速原型 | 團隊協作、框架設計 |
| Python 風格 | 非常 Pythonic | 較接近 Java 風格 |
| 使用複雜度 | 簡單 | 需要額外 import |

**簡單判斷法：**
- 自己寫小程式 → 鴨子型別就夠了
- 設計讓別人實作的介面 → 用 ABC，保護大家不犯錯

## 5. 實用技巧：用 ABC 定義協定

```python
from abc import ABC, abstractmethod

class Exportable(ABC):
    """任何可匯出的物件都要實作這個介面"""

    @abstractmethod
    def to_dict(self):
        pass

    def to_json(self):
        import json
        return json.dumps(self.to_dict(), ensure_ascii=False)
```

這樣只要繼承 `Exportable` 並實作 `to_dict()`，就自動獲得 `to_json()` 的能力。

## 重點整理

| 概念 | 說明 |
|---|---|
| 多型 | 同一個介面，不同物件有不同行為 |
| 鴨子型別 | Python 只看「有沒有這個方法」，不看型別 |
| ABC | 抽象基底類別，強制子類別實作指定方法 |
| `@abstractmethod` | 標記必須被子類別覆寫的方法 |
| 抽象類別不能實例化 | 只能當作「藍圖」被繼承 |
