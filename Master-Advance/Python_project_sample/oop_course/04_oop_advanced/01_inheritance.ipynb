## 1. 什麼是繼承？

繼承就像家族關係：孩子會從父母那裡「繼承」某些特徵，同時也有自己的獨特之處。

在程式中，**子類別（child class）** 會自動擁有 **父類別（parent class）** 的所有屬性和方法，
同時可以新增或修改自己的行為。

```
父類別（Parent） ──繼承──> 子類別（Child）
  共通的屬性與方法          新增或覆寫的部分
```

## 2. 基本語法

````python
class Animal:
    """父類別：動物"""
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} 發出聲音"

class Dog(Animal):
    """子類別：狗，繼承自 Animal"""
    def speak(self):
        return f"{self.name} 說：汪汪！"

class Cat(Animal):
    """子類別：貓，繼承自 Animal"""
    def speak(self):
        return f"{self.name} 說：喵～"
````

````python
dog = Dog("小黑")
cat = Cat("橘子")

print(dog.speak())  # 小黑 說：汪汪！
print(cat.speak())  # 橘子 說：喵～
print(dog.name)     # 小黑 （繼承自 Animal 的屬性）
````

重點：`Dog` 和 `Cat` **沒有定義 `__init__`**，但自動繼承了 `Animal` 的 `__init__`。

## 3. 方法覆寫（Override）

子類別可以重新定義父類別的方法，這稱為「覆寫」：

````python
class Animal:
    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):          # 覆寫父類別的 speak
        return "汪汪！"
````

當你呼叫 `Dog().speak()` 時，Python 會先在 `Dog` 裡找，找到了就用子類別的版本。

## 4. super() 的使用

`super()` 讓你在子類別中呼叫父類別的方法，最常見的用途是在 `__init__` 中：

````python
class Employee:
    def __init__(self, name, employee_id):
        self.name = name
        self.employee_id = employee_id

    def describe(self):
        return f"{self.name} (編號: {self.employee_id})"

class Manager(Employee):
    def __init__(self, name, employee_id, department):
        super().__init__(name, employee_id)   # 先呼叫父類別的 __init__
        self.department = department           # 再設定自己的屬性

    def describe(self):
        base = super().describe()              # 重用父類別的方法
        return f"{base}, 部門主管: {self.department}"

class Engineer(Employee):
    def __init__(self, name, employee_id, language):
        super().__init__(name, employee_id)
        self.language = language

    def describe(self):
        base = super().describe()
        return f"{base}, 專長: {self.language}"
````

````python
mgr = Manager("Alice", "M001", "研發部")
eng = Engineer("Bob", "E042", "Python")

print(mgr.describe())  # Alice (編號: M001), 部門主管: 研發部
print(eng.describe())  # Bob (編號: E042), 專長: Python
````

### 什麼時候要呼叫 super()？

| 情境 | 要不要呼叫 super()？ |
|---|---|
| 子類別的 `__init__` 需要父類別的初始化邏輯 | 要 |
| 想在覆寫方法中保留父類別的行為再擴充 | 要 |
| 完全取代父類別的方法，不需要原本邏輯 | 不需要 |

## 5. isinstance() 檢查繼承關係

````python
print(isinstance(mgr, Manager))   # True
print(isinstance(mgr, Employee))  # True （Manager 也是 Employee）
print(isinstance(eng, Manager))   # False
````

## 6. 常見陷阱：不要濫用繼承

繼承表達的是 **「is-a」** 關係。如果關係不成立，就不該用繼承：

````python
# 錯誤示範：Stack 不是 list，不應繼承
class Stack(list):
    def push(self, item):
        self.append(item)
# 問題：使用者可以呼叫 insert、sort 等不該有的方法

# 正確做法：使用組合（composition）
class Stack:
    def __init__(self):
        self._items = []       # 把 list 當成內部元件

    def push(self, item):
        self._items.append(item)

    def pop(self):
        return self._items.pop()

    def is_empty(self):
        return len(self._items) == 0
````

**判斷原則：**
- "Dog is an Animal" → 用繼承
- "Car has an Engine" → 用組合（把 Engine 當屬性）

## 重點整理

| 概念 | 說明 |
|---|---|
| 繼承 | 子類別自動擁有父類別的屬性與方法 |
| 覆寫（Override） | 子類別重新定義父類別的方法 |
| `super()` | 在子類別中呼叫父類別的方法 |
| `isinstance(obj, Class)` | 檢查物件是否屬於某類別（含繼承） |
| 組合優於繼承 | 「has-a」關係用組合，「is-a」關係用繼承 |
