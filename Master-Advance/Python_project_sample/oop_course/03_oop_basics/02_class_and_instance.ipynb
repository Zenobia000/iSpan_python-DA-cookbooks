## 類別就是藍圖

蓋房子之前，你需要一張設計圖。設計圖本身不是房子，但每蓋一棟房子都按照它來建。

```
類別 (class)  = 設計圖     → 定義結構和行為
實例 (instance) = 蓋出來的房子 → 每棟都獨立存在
```

---

## 第一個類別

````python
class Dog:
    def __init__(self, name, breed):
        self.name = name      # 實例屬性
        self.breed = breed    # 實例屬性

    def bark(self):
        return f"{self.name} 說：汪汪！"

# 建立兩個實例
d1 = Dog("小白", "柴犬")
d2 = Dog("旺財", "拉布拉多")

print(d1.bark())  # 小白 說：汪汪！
print(d2.bark())  # 旺財 說：汪汪！
print(d1.name)    # 小白
print(d2.name)    # 旺財
````

### 逐行解說

| 程式碼 | 意思 |
|--------|------|
| `class Dog:` | 定義一個叫 Dog 的類別 |
| `def __init__(self, name, breed):` | 建構式，建立實例時自動執行 |
| `self.name = name` | 把參數存到實例身上 |
| `def bark(self):` | 定義一個方法（屬於實例的函式） |
| `Dog("小白", "柴犬")` | 建立一個 Dog 實例 |

## self 是什麼？

`self` 就是「這個實例自己」。Python 會自動傳入，你不需要手動給。

````python
class Cat:
    def __init__(self, name):
        self.name = name      # self.name 是屬性，name 是參數

    def introduce(self):
        return f"我是 {self.name}"  # 透過 self 取得自己的資料

c = Cat("咪咪")
c.introduce()   # Python 自動把 c 傳給 self
# 等同於 Cat.introduce(c)
````

> **記法：** 在方法裡要存取實例的東西，就用 `self.xxx`。

## 實例屬性 vs 類別屬性

````python
class Student:
    school = "大同大學"       # 類別屬性：所有學生共用

    def __init__(self, name, grade):
        self.name = name      # 實例屬性：每個學生不同
        self.grade = grade

s1 = Student("小明", 85)
s2 = Student("小華", 92)

# 類別屬性：所有實例共享同一個值
print(s1.school)  # 大同大學
print(s2.school)  # 大同大學

# 實例屬性：各自獨立
print(s1.name)    # 小明
print(s2.name)    # 小華
````

### 類別屬性的陷阱

````python
# 透過實例「修改」類別屬性，其實是建立了新的實例屬性
s1.school = "台大"
print(s1.school)         # 台大（s1 自己的實例屬性）
print(s2.school)         # 大同大學（還是類別屬性）
print(Student.school)    # 大同大學（類別屬性沒變）

# 要真正改類別屬性，要透過類別本身
Student.school = "成大"
print(s2.school)         # 成大
````

## 屬性查找順序

Python 找屬性時，會先找實例、再找類別：

```
obj.xxx
  1. 先找 obj 自己有沒有 xxx（實例屬性）
  2. 沒有的話，找 obj 的類別有沒有 xxx（類別屬性）
  3. 都沒有 → AttributeError
```

---

## 完整範例：Student 類別

````python
class Student:
    passing_score = 60    # 類別屬性

    def __init__(self, name, scores=None):
        self.name = name
        self.scores = list(scores) if scores else []

    def add_score(self, score):
        if not 0 <= score <= 100:
            raise ValueError(f"分數 {score} 超出範圍 0~100")
        self.scores.append(score)

    def average(self):
        if not self.scores:
            return 0
        return sum(self.scores) / len(self.scores)

    def is_passing(self):
        return self.average() >= self.passing_score

    def summary(self):
        status = "及格" if self.is_passing() else "不及格"
        return f"{self.name}：平均 {self.average():.1f} ({status})"

# 使用
s = Student("小明", [85, 72, 90])
s.add_score(88)
print(s.summary())  # 小明：平均 83.8 (及格)
````

## 練習題

試著寫一個 `Book` 類別：

1. 屬性：`title`（書名）、`author`（作者）、`pages`（頁數）
2. 方法 `is_long()`：超過 300 頁回傳 `True`
3. 方法 `info()`：回傳 `"《書名》by 作者 (N頁)"`

````python
# 預期結果
b = Book("Python 入門", "王小明", 250)
print(b.info())       # 《Python 入門》by 王小明 (250頁)
print(b.is_long())    # False
````

<details>
<summary>參考解答</summary>

````python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def is_long(self):
        return self.pages > 300

    def info(self):
        return f"《{self.title}》by {self.author} ({self.pages}頁)"
````

</details>

## 本節重點

| 概念 | 說明 |
|------|------|
| 類別 (class) | 定義物件的結構與行為的藍圖 |
| 實例 (instance) | 根據類別建立的具體物件 |
| `__init__` | 建構式，建立實例時自動執行 |
| `self` | 代表實例本身，方法的第一個參數 |
| 實例屬性 | 每個實例各自擁有，在 `__init__` 中用 `self.xxx` 建立 |
| 類別屬性 | 所有實例共用，直接寫在 class 區塊內 |
