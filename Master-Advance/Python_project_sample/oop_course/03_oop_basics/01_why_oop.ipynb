## 一個簡單的問題：學生成績系統

假設我們要管理學生資料，包含姓名、分數，並計算等級。

### 函式導向的寫法

````python
# 用 dict 存資料，用函式處理邏輯
def create_student(name, scores):
    return {"name": name, "scores": scores}

def average(student):
    return sum(student["scores"]) / len(student["scores"])

def grade(student):
    avg = average(student)
    if avg >= 90: return "A"
    if avg >= 80: return "B"
    if avg >= 70: return "C"
    return "F"

# 使用
s1 = create_student("小明", [85, 92, 78])
print(f"{s1['name']} 平均：{average(s1):.1f}，等級：{grade(s1)}")
````

這樣寫完全可以運作，但隨著需求成長，問題會浮現。

### 問題在哪？

````python
# 任何人都能直接改資料，沒有任何保護
s1["scores"].append(-100)   # 不合理的分數，沒人阻止
s1["name"] = ""             # 空名字也沒問題？

# 函式和資料是分開的，你必須記得哪些函式搭配哪種 dict
average({"x": 1})  # 執行才會爆錯，IDE 幫不了你
````

### 物件導向的寫法

````python
class Student:
    def __init__(self, name, scores):
        self.name = name
        self.scores = list(scores)  # 複製一份，避免外部修改

    def average(self):
        return sum(self.scores) / len(self.scores)

    def grade(self):
        avg = self.average()
        if avg >= 90: return "A"
        if avg >= 80: return "B"
        if avg >= 70: return "C"
        return "F"

    def add_score(self, score):
        if not 0 <= score <= 100:
            raise ValueError("分數必須在 0~100 之間")
        self.scores.append(score)

# 使用
s1 = Student("小明", [85, 92, 78])
print(f"{s1.name} 平均：{s1.average():.1f}，等級：{s1.grade()}")

s1.add_score(95)    # OK
s1.add_score(-100)  # ValueError: 分數必須在 0~100 之間
````

**關鍵差異：** 資料和行為綁在一起，還能加上驗證保護。

## 兩種方式並排比較

```
# --- 函式導向 ---                    # --- 物件導向 ---
s = create_student("小明", [85])      s = Student("小明", [85])
avg = average(s)                      avg = s.average()
s["scores"].append(-1)  # 沒保護      s.add_score(-1)  # 會擋下來
```

## 什麼時候用什麼？決策樹

```
你的問題有沒有「東西」需要管理？
│
├─ 沒有 → 純計算、轉換、工具函式
│         例如：計算 BMI、轉換日期格式
│         >>> 用函式就好
│
└─ 有 → 這些「東西」有狀態（資料）嗎？
         │
         ├─ 沒有 → 只是把相關函式分組
         │         >>> 用模組 (module) 即可
         │
         └─ 有 → 需要多個同類的東西嗎？
                  │
                  ├─ 只有一個 → dict 或 dataclass 就夠了
                  │
                  └─ 多個，且行為複雜 → 用 class
                           例如：多個學生、多個帳戶、遊戲角色
```

## 常見誤解

| 誤解 | 事實 |
|------|------|
| OOP 比較「高級」 | 兩者是不同工具，沒有高低之分 |
| 什麼都該用 class | 簡單任務用函式更清楚 |
| Python 一定要 OOP | Python 支援多種風格，混用很正常 |
| OOP 一定比較慢 | 效能差異在多數場景可忽略 |

---

## 本節重點

| 項目 | 函式導向 | 物件導向 |
|------|----------|----------|
| 資料與行為 | 分開管理 | 綁在一起 |
| 資料保護 | 需自行注意 | 可透過封裝控制 |
| 適合場景 | 簡單計算、工具函式 | 有狀態的複雜物件 |
| 程式碼組織 | 靠函式名稱與模組 | 靠類別結構 |
| 學習門檻 | 較低 | 稍高，但長期受益 |
