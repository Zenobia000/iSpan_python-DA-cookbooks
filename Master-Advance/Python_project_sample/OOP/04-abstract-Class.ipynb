##  Animal

 class    python abc  ABCMeta <br>
ABCMeta  Python abc Abstract Base Class ABCMeta 

```python
from abc import ABCMeta, abstractmethod, ABC


# :

#  1
# class Animal(ABC):
#     @abstractmethod
#     def make_sound(self):
#         pass

#  2
class Animal(metaclass=ABCMeta):
    @abstractmethod
    def make_sound(self):
        pass
```

## 

```python
class Dog(Animal):
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print(f"{self.name} said Woof!")

class Cat(Animal):
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print(f"{self.name} said Meow!")

class Cow(Animal):
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print(f"{self.name} said Moo!")

```

```python
dog = Dog("jake")
dog.make_sound()  # : Woof!

cat = Cat("wendy")
cat.make_sound()  # : Meow!

cow = Cow("bob")
cow.make_sound()  # : Moo!
```

>  interface  for loop 

```python
for animal in [dog, cat, cow]:
    animal.make_sound()
```

```python
from abc import ABCMeta, abstractmethod

# 
class Shape(metaclass=ABCMeta):
    @abstractmethod
    def draw(self):
        pass

    @abstractmethod
    def area(self):
        pass

# 
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def draw(self):
        print(f"Drawing a circle with radius {self.radius}")
    
    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def draw(self):
        print(f"Drawing a rectangle with width {self.width} and height {self.height}")
    
    def area(self):
        return self.width * self.height

# 
circle = Circle(5)
circle.draw()  # : Drawing a circle with radius 5
print(circle.area())  # : 78.5

rectangle = Rectangle(4, 6)
rectangle.draw()  # : Drawing a rectangle with width 4 and height 6
print(rectangle.area())  # : 24

```
