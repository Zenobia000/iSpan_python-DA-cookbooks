# 長條圖 bar()
- 通常用於「類別型」資料 (categorical data)
  - 資料之間是離散的 (discrete)、間斷的
- plt.bar(x, height, width=0.8, ...)
  - `x`: array-like 的資料，例如 list、array、tuple 等，作為 x 軸的標題
  - `height`: array-like 的資料，作為 y 軸的值
  - `width`: bar 的寬度
- matplotlib.pyplot.xticks(ticks=None, labels=None, ...)
  - `ticks`: x 軸的刻度
  - `label`: 對應 x 軸的標題，類似取代 x 軸的刻度，用其它名稱代替
- 參考連結:
  - [matplotlib.pyplot.bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)
  - [matplotlib.pyplot.xticks](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html)
  - [Python matplotlib 畫圖 – 柱狀圖 (長條圖)](https://python-ecw.com/2020/11/12/matplotlib柱狀圖/)

```python
'''
班上投票，Alex 得票 7749，Bill 得票 9527，Carl 得票 2266，請用長條圖表示
'''

# 得票數 (因為有 3 名候選人)
votes = [7749, 9527, 2266]
candidates = ['Alex', 'Bill', 'Carl']

# x 軸的刻度，例如 [0, 1, 2]
x_index = np.arange(len(votes))

# bar 的寬度
width = 0.5

# 繪製長條圖
plt.bar(x_index, votes, width)

# 設定數值文字在圖表上
for i in range(len(x_index)):
    plt.text(x_index[i], votes[i], votes[i])

# 設定圖表標題
plt.title('Election')
plt.ylabel('Votes')

# 設定 x 和 y 軸的刻度
plt.xticks(x_index, candidates)
plt.yticks(np.arange(0, 10000, 1000))

# 顯示圖表
plt.show()
```

# 直方圖
- 通常用於「數值型」資料 (numeric data)
  - 資料之間是連續的 (continueous)
- plt.hist(x, bins, ...)
  - x: 你的資料，list、array 或是 sequece
  - bin: 組距，預設 10
- 參考資料:
  - [matplotlib.pyplot.hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)
  - [Python 學習筆記 : Matplotlib 資料視覺化 (二) 統計圖](http://yhhuang1966.blogspot.com/2020/05/python-matplotlib.html)

```python
'''
請以班上 100 個學生考試成績，最低 0 分，最高 100 分，繪製直方圖
'''

# 隨機產生 100 位同學的成績 (用 np.random.randint 的話，要轉成 list)
x_scores = list(np.random.randint(0, 100, size=(1, 100), dtype='int32'))

# 組距
x_bins = np.arange(0, 101, 10)

# 繪製直方圖 (會回傳每個組距的數量，以及對應組距)
arr_counts, arr_bins, obj = plt.hist(x_scores, bins=x_bins)

# 設定數值文字在圖表上
for i in range(len(arr_counts)):
    plt.text(arr_bins[i], arr_counts[i], arr_counts[i].astype('int32'))

# 設定標題
plt.title('Students Exam')
plt.xticks(x_bins, x_bins)
plt.ylabel('Scores')

# 顯示圖表
plt.show()
```

# 圓餅圖 pie()
- 用於描述「類別型」資料
- plt.pie(x, explode=None, labels=None, colors=None, autopct=None,shadow=False, labeldistance=1.1, radius=1, counterclock=True, center=(0, 0), ...)
  - x: 你的資料。
  - explode: 餅分離的程度，list 格式，其中的每個元素值介於 0 到 1 之間。
  - labels: 圓餅圖的標題，list 格式。
  - autopct: 每一個項目的百分比格式，例如「%2.2f%%」，代表整數 2 位數，小數 2 位數。
  - labeldistance: 項目標題與圓餅圖中間距離是半徑的多少倍。
  - center: 圓中心座標，預設 (0, 0)。
  - shadow: 圓餅圖是否有陰影，預設為 False。
- 參考資料:
  - [matplotlib.pyplot.pie](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie)

```python
'''
有一家庭支出費用如下，以此設計圓餅圖:
旅行 (Travel): 8000
娛樂 (Entertainment): 2000
教育 (Education): 3000
交通 (Transportation): 5000
餐費 (Food): 6000
'''

# 設定支出項目
items = ['Travel', 'Entertainment', 'Education', 'Transportation', 'Food']

# 設定支出費用
expenses = [8000, 2000, 3000, 5000, 6000]

# 繪製圓餅圖
plt.pie(
    expenses,
    labels=items,
    explode=(0,0.2,0,0,0),
    autopct="%1.2f%%"
)

# 顯示圖表
plt.show()
```

# 其它範例

```python
'''Stacked Bar chart'''

# 匯入套件
import matplotlib.pyplot as plt
import numpy as np

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

# 建立資料
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
 
# 繪製 stacked bar
# (重點在 bottom 屬性; y1 和 y2 y3 可以相加，是因為 numpy)
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')

# 圖表標題
plt.xlabel("隊伍")
plt.ylabel("分數")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("每個隊伍在 4 輪中的分數")

# 顯示圖表
plt.show()
```

```python
'''
Horizontal Bar chart
'''
# 匯入套件
import matplotlib.pyplot as plt

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

# 設定類別項目
tick_labels = ['Python', 'Scala', 'C#', 'Java', 'PHP']

# 設定類別項目的 x 軸順序
x_axis_index = [1,2,3,4,5]

# 每個類別項目的值
y = [50, 10, 13, 27, 22]

# 繪製 Horizontal Bar
plt.barh(x_axis_index, y, tick_label=tick_labels)

# 設定數值文字在圖表上
for i in range(len(x_axis_index)):
    plt.text(y[i], x_axis_index[i], y[i])

# 圖表標題
plt.ylabel("使用次數")
plt.xlabel("程式語言")

# 顯示圖表
plt.show()
```

```python
'''
Grouped Bar chart
'''
# 匯入套件
import matplotlib.pyplot as plt

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

# 設定群組資料的值 (第1隊、第2隊)
y_score_1 = [35, 60, 18, 29, 77]
y_score_2 = [28, 33, 93, 44, 32]

# 設定 x 軸的順序，這裡要注意的是，group 的 bar chart，要另外指定 x 軸的位置
BAR_WIDTH = 0.4
x_axis_index1 = [1, 2, 3 , 4, 5]
x_axis_index2 = [i + BAR_WIDTH for i in x_axis_index1]
'''
x_axis_index2 = [
    1 + BAR_WIDTH,
    2 + BAR_WIDTH,
    3 + BAR_WIDTH,
    4 + BAR_WIDTH,
    5 + BAR_WIDTH
]
'''

# 設定 x 刻度的中心位置
ticks = [i + BAR_WIDTH / 2 for i in x_axis_index1]
'''
ticks = [
    1 + BAR_WIDTH / 2,
    2 + BAR_WIDTH / 2,
    3 + BAR_WIDTH / 2,
    4 + BAR_WIDTH / 2,
    5 + BAR_WIDTH / 2
]
'''

# 設定 x 標題
tick_labels = ["Round 1", "Round 2", "Round 3", "Round 4", "Round 5"]

# 繪製圖表
plt.bar(x_axis_index1, y_score_1, BAR_WIDTH, color="y", label="第 1 隊")
plt.bar(x_axis_index2, y_score_2, BAR_WIDTH, color="b", label="第 2 隊")

# 設定數值文字在圖表上
for i in range(len(x_axis_index1)):
    plt.text(x_axis_index1[i], y_score_1[i], y_score_1[i])
for i in range(len(x_axis_index2)):
    plt.text(x_axis_index2[i], y_score_2[i], y_score_2[i])

# 圖表標題
plt.xlabel("回合數")
plt.ylabel("分數")
plt.title("隊伍競賽分數")
plt.xticks(ticks, tick_labels)
plt.legend(loc='upper left', ncols=2)

# 顯示圖表
plt.show()
```
